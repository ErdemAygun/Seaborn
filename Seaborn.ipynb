### Heatmap Chart

```python
df = sns.load_dataset("flights")
df
```

```python
df.dtypes
```

```python
df["year_str"] = df.year.astype('object') #df.year.apply(lambda s:str(s) )
df["month_str"] = df.month.astype('object') #df.month.apply(lambda s:"_"+str(s) )
df
```

```python
df.dtypes
```

```python
df["year_month"] = df["year"].astype('str') + "_" + df["month"].astype('str') #df.year + " " + df.month
```

```python
df
```

```python
df_temp = pd.pivot_table(df, index="month", columns="year", values="passengers")
df_temp
```

```python
plt.figure(figsize = (9,7))
sns.heatmap(df_temp)
plt.show()
```

```python
plt.figure(figsize = (9,7))
sns.heatmap(df_temp, annot=True)
plt.show()
```

```python
plt.figure(figsize = (9,7))
sns.heatmap(df_temp, annot=True, fmt="d")
plt.show()
```

```python
plt.figure(figsize = (9,7))
sns.heatmap(df_temp, annot=True, fmt="d", cmap="jet")
plt.show()
```

```python
plt.figure(figsize = (9,7))
sns.heatmap(df_temp, annot=True, fmt="d", cmap="jet", linewidth=1)
plt.show()
```
